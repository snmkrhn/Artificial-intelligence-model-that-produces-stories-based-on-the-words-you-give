```python
import re
```

```python
# Elle stopwords listesi oluşturalım (İngilizce için temel durak kelimeler)
stopwords = ['the', 'and', 'is', 'in', 'to', 'of', 'a', 'that', 'it', 'on', 'for', 'with', 'as', 'by', 'at', 'an', 'this', 'be', 'from', 'or']

# Temizlik fonksiyonunu güncelleyelim
def clean_text(text):
    if isinstance(text, str):  # Sadece string tipindeki değerlere işlem yap
        text = text.lower()  # Tüm metni küçük harfe çevirme
        text = re.sub(r'\d+', '', text)  # Sayıları kaldırma
        text = re.sub(r'[^\w\s]', '', text)  # Noktalama işaretlerini kaldırma
        text = ' '.join([word for word in text.split() if word not in stopwords])  # Durak kelimeleri çıkarma
        return text
    return text  # Eğer metin değilse (örneğin NaN veya sayı), olduğu gibi döndür

# Temizleme fonksiyonunu veri çerçevesine uygulama
df['cleaned_text'] = df['text'].apply(clean_text)

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# Veri setini yükleyin
df = pd.read_csv('/kaggle/input/tinystories-narrative-classification/train.csv')

# Eksik verileri temizleyin
df = df.dropna(subset=['text'])

# Veri setinin küçük bir örneğiyle çalışın (örneğin, ilk 1000 satır)
df_sample = df.sample(1000)

# Metin verisini içeren sütunu alın
texts = df_sample['text'].values

# Metin verisini sayısal verilere dönüştürün
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

# NMF modelini oluşturun
nmf_model = NMF(n_components=5, random_state=0)
nmf_model.fit(X)

# Konu terimlerini inceleyin
for index, topic in enumerate(nmf_model.components_):
    print(f"Topic #{index}:")
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])

```

```python
import numpy as np

# Her bir hikaye için NMF modelinin konu dağılımlarını al
topic_distributions = nmf_model.transform(X)

# En yüksek olasılığa sahip olan konuyu her hikaye için etiket olarak ata
df_sample['topic_label'] = np.argmax(topic_distributions, axis=1)

# Etiketlenmiş veriyi kontrol edelim
print(df_sample[['text', 'topic_label']].head())
```

```python
# İlk 10 satırı görmek için
print(df_sample[['text', 'topic_label']].head(10))

# Belirli bir konuya ait hikayeleri görmek için (örneğin, 'topic_label' 2 olanlar)
print(df_sample[df_sample['topic_label'] == 2][['text', 'topic_label']])

# Tüm etiketi olan veriyi CSV dosyasına kaydetmek istersen:
df_sample.to_csv('etiketli_veri.csv', index=False)
```

```python
# Topic Label 0 olan hikayeleri gör
topic_0_stories = df_sample[df_sample['topic_label'] == 0]
print(topic_0_stories[['text', 'topic_label']])

# Topic Label 1 olan hikayeleri gör
topic_1_stories = df_sample[df_sample['topic_label'] == 1]
print(topic_1_stories[['text', 'topic_label']])
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Text verisini hazırlayın (girdi verisi olarak kullanacağız)
texts = df_sample['text'].values

# Tokenizer oluşturun
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

# Metinleri sayısal verilere dönüştürün (tokenizasyon)
sequences = tokenizer.texts_to_sequences(texts)

# Verileri aynı uzunluğa getirin (padding işlemi)
maxlen = 100  # Giriş uzunluğu olarak 100 seçtim, bunu ihtiyaca göre artırabilirsiniz
X = pad_sequences(sequences, maxlen=maxlen)

# Konu etiketlerini hedef değerler (output) olarak kullanacağız
y = df_sample['topic_label'].values

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

# Modeli oluşturun
model = Sequential()

# Embedding katmanı (her kelimeyi bir vektör temsil edecek)
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=50))

# RNN katmanı
model.add(SimpleRNN(128, return_sequences=False))

# Çıkış katmanı (5 farklı konu olduğu için)
model.add(Dense(5, activation='softmax'))

# Modeli derleyin
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Modeli eğitin
model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)

```

```python
# Yeni bir kelime girdisi verelim
new_input = ['dog']

# Tokenizer kullanarak yeni girdiyi sayısal forma dönüştürelim
new_sequence = tokenizer.texts_to_sequences(new_input)
new_sequence = pad_sequences(new_sequence, maxlen=maxlen)

# Modelin tahminini alalım
predicted_topic = model.predict(new_sequence)

# En yüksek olasılıkla tahmin edilen konuyu belirleyelim
predicted_label = np.argmax(predicted_topic)

# O konuya ait bir hikaye seçip yazdıralım
predicted_story = df_sample[df_sample['topic_label'] == predicted_label].sample(1)['text'].values[0]
print(f"Tahmin edilen konu: {predicted_label}, Hikaye: {predicted_story}")
```

```python
# Üç kelimelik yeni bir girdi verelim
new_input = ['dog', 'play', 'park']

# Tokenizer kullanarak yeni girdiyi sayısal forma dönüştürelim
new_sequence = tokenizer.texts_to_sequences([' '.join(new_input)])
new_sequence = pad_sequences(new_sequence, maxlen=maxlen)

# Modelin tahminini alalım
predicted_topic = model.predict(new_sequence)

# En yüksek olasılıkla tahmin edilen konuyu belirleyelim
predicted_label = np.argmax(predicted_topic)

# O konuya ait bir hikaye seçip yazdıralım
predicted_story = df_sample[df_sample['topic_label'] == predicted_label].sample(1)['text'].values[0]
print(f"Tahmin edilen konu: {predicted_label}, Hikaye: {predicted_story}")
```
